```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import skfuzzy as fuzz
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
```

```python
data = X.T
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
    data, c=3, m=2, error=0.005, maxiter=1000, init=None
)
print("Cluster centers:\n", cntr)
print("Fuzzy Partition Coefficient (FPC):", fpc)
```

```python
cluster_labels = np.argmax(u, axis=0)
plt.figure(figsize=(7,5))
for c in range(3):
    plt.scatter(X[cluster_labels == c, 0], X[cluster_labels == c, 1], label=f'Cluster {c}')
plt.scatter(cntr[:, 0], cntr[:, 1], marker='x', s=200, linewidths=3, color='black', label='Centers')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Fuzzy C-Means Clustering on Iris Dataset")
plt.legend()
plt.show()
```

## Conclusion

Fuzzy C-Means clustering was applied to the Iris dataset.
The results show the division into three fuzzy clusters, with the FPC value indicating the quality of the partition.
